```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df_final=df_all_rows[['distance','Venue Latitude','Venue Longitude','FoodType','Neighborhood']]
df_final['Neighborhood']=le.fit_transform(df_final['Neighborhood'])
df_final['FoodType']=le.fit_transform(df_final['FoodType'])
```

```python
from sklearn.metrics import silhouette_score
sil=[]
kmax=25
for k in range(2,kmax+1):
  kmeans=KMeans(n_clusters=k).fit(df_final)
  labels=kmeans.labels_
  sil.append(silhouette_score(df_final,labels,metric="euclidean"))
```

```python
kclusters = 3

df_final=df_all_rows[['Venue Latitude','Venue Longitude','FoodType']]
df_final['FoodType']=le.fit_transform(df_final['FoodType'])
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_final)
df_final['Cluster']=kmeans.labels_
df_final['Cluster']=df_final['Cluster'].apply(str)

map=folium.Map(location=[12.8406,80.1534],zoom_start=12)    #12.8406° N, 80.1534° E

# instantiate a feature group for the incidents in the dataframe
locations = folium.map.FeatureGroup()

# set color scheme for the clusters
def color_producer(cluster):
    if cluster=='0':
        return 'green'
    elif cluster=='1':
        return 'blue'
    else:
        return 'red'

latitudes = list(df_final['Venue Latitude'])
longitudes = list(df_final['Venue Longitude'])
labels = list(df_final['Cluster'])
names=list(df_all_rows['Venue'])
distance=list(df_all_rows['distance'])
cuisine=list(df_all_rows['FoodType'])
distance_from=list(df_all_rows['Neighborhood'])
address=list(df_all_rows['address'])


for lat, lng, label,names,distance,cuisine,distance_from,address in zip(latitudes, longitudes, labels,names,distance,cuisine,distance_from,address):
    my_string = 'Name: {}, cuisine: {},address: {}, distance from map location to entrance: {}'.format(names,cuisine,address,distance)
    folium.Marker(
            [lat,lng],
            fill=True,
            fill_opacity=1,
            popup=folium.Popup(my_string,max_width=300),
            radius=5,
            color=color_producer(label)
        ).add_to(map)
    folium.CircleMarker(
            [lat,lng],
            fill=True,
            fill_opacity=1,
            #popup=folium.Popup(my_string,max_width=300),
            radius=7,
            color=color_producer(label)
        ).add_to(map)

# add locations to map
map.add_child(locations)
folium.Marker([19.1334,72.9133],popup='VIT Chennai').add_to(map)

#saving the map 
```
